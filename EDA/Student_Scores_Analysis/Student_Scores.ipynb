## Drop unnamed column

```python
df = df.drop('Unnamed: 0', axis=1)
df.head(10)
```

## Gender distribution

```python
plt.figure(figsize=(7, 5))
ax= sns.countplot(data = df, x = 'Gender', hue='Gender')
plt.title('Gender Distribution')
for bars in ax.containers:
    ax.bar_label(bars)
```

# **Finding:**
***From the above chart we have analysed that the number of females in the data is more than the number of males.***

```python
gb = df.groupby('ParentEduc').agg({"MathScore": "mean", "ReadingScore": "mean", "WritingScore": "mean"})
print(gb)
```

```python
sns.heatmap(gb, annot=True)
plt.title("Relationship between Parent's Education and Student's scores")
plt.show()
```

# **Finding:**
***From the above chart we have concluded that the education of the parents have a good impact on student's scores.***

```python
gb1 = df.groupby('ParentMaritalStatus').agg({"MathScore": "mean", "ReadingScore": "mean", "WritingScore": "mean"})
print(gb1)
```

```python
sns.heatmap(gb1, annot=True)
plt.title("Relationship between Parent's marital status and Student's scores")
plt.show()
```

# **Finding:**
***From the above chart we have concluded that there is no/negligible impact on the student's scores due to their parent's marital status.***

## Distribution of Ethnic Groups

```python
groupA = df.loc[df["EthnicGroup"] == "group A"].count()
groupB = df.loc[df["EthnicGroup"] == "group B"].count()
groupC = df.loc[df["EthnicGroup"] == "group C"].count()
groupD = df.loc[df["EthnicGroup"] == "group D"].count()
groupE = df.loc[df["EthnicGroup"] == "group E"].count()

label = ["group A", "group B", "group C", "group D", "group E"]
myList = [groupA["EthnicGroup"], groupB["EthnicGroup"], groupC["EthnicGroup"], groupD["EthnicGroup"], groupE["EthnicGroup"]]
plt.pie(myList, labels=label, autopct='%1.2f%%')
plt.title("Distribution of Ethnic Groups")
plt.show()
```
