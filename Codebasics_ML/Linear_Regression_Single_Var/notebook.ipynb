# **Linear regression single variable**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

### **Read the CSV file**

```python
df =  pd.read_csv("canada_per_capita_income.csv")
df.head()
```

### **Plot the year & per_capita_incode**

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per_capita_income')
plt.scatter(df.year, df.per_capita_income, color='blue', marker='*')
plt.scatter(df.year, df.per_capita_income, color='red', marker='+')
```

### **Use Linear Model & fit the data**

```python
reg = linear_model.LinearRegression()
reg.fit(df[['year']], df.per_capita_income)
```

### **Predict using the linear model**

```python
reg.predict([[2017]])
```

### **Internal details about this prediction**

```python
# y = m * x + b
# m = coeficient
# b = intercept
# x = year

m = reg.coef_
x = 2017
b = reg.intercept_
y = m * x + b

print(y)
```

### **Predict per_capita_income for multiple year in year.csv and create a new colum to shoe the per_capita_income**

```python
df_1 = pd.read_csv("year.csv")
df_1.head()
```

```python
prediction_1 = reg.predict(df_1)
df_1['per_capita_income'] = prediction_1
```
